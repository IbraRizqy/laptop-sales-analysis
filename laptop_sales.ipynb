Data Cleaning

```python
# Memisah data untuk produk dan Laptop
df_phones = df[df['Product'] == 'Mobile Phone']
df_phones = df_phones.drop(columns=['Core Specification', 'SSD', 'Product'])

df_phones.to_excel('phones_sales_data.xlsx', index=False)

df_laptops = df[df['Product'] == 'Laptop']
df_laptops.drop(columns=['Product', 'Processor Specification'], inplace=True)
df_laptops.info()
```

Membuat kolom baru

```python
# Membuat Total Revenue dan Inventory Time
df_laptops['Total Revenue'] = df['Price'] * df['Quantity Sold']
df['Inward Date'] = pd.to_datetime(df_laptops['Inward Date'])
df['Dispatch Date'] = pd.to_datetime(df_laptops['Dispatch Date'])
df_laptops['Inventory Time'] =  df['Dispatch Date'] - df['Inward Date']
```

Training untuk kolom Performance Rating

```python
print(df_laptops['Core Specification'].unique())
print(df_laptops['RAM'].unique())
print(df_laptops['SSD'].unique())
```

```python
#Mapping core specifications, RAM, and SSD ke numerik
core_spec_mapping = {
    'i3': 1,
    'i5': 2,
    'i7': 3,
    'i9': 4,
    'Ryzen 3': 1,
    'Ryzen 5': 2,
    'Ryzen 7': 3,
    'Ryzen 9': 4
}
ram_mapping = {
    '4GB': 4,
    '6GB': 6,
    '8GB': 8,
    '12GB': 12,
    '16GB': 16,
    '32GB': 32
}
ssd_mapping = {
    '256GB': 256,
    '512GB': 512,
    '1TB': 1024,
    '2TB': 2048
}
df_laptops['Core Specification'] = df_laptops['Core Specification'].map(core_spec_mapping)
df_laptops['RAM'] = df_laptops['RAM'].map(ram_mapping)
df_laptops['SSD'] = df_laptops['SSD'].map(ssd_mapping)
```

```python
# Training dengan KMeans
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = df_laptops[['Core Specification', 'RAM', 'SSD']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features_scaled)

labels_mapping = {0: 'Low', 1: 'Medium', 2: 'High'}
df_laptops['Performance Rating'] = kmeans.labels_
df_laptops['Performance Rating'] = df_laptops['Performance Rating'].map(labels_mapping)
```
